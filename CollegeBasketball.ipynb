```python
import numpy as np
import pandas as pd

data = pd.read_excel("NCAA 2020-21 - Top 50.xlsx"); 
```

```python
data = data.set_index("School")
data
```

```python

top_25_teams = data.dropna(axis = 0)
top_25_teams
```

# FG Percent Correlation

```python
import scipy.stats as stats
import re

percent_correlation_FG_percent = stats.pearsonr(top_25_teams['Associated Press Ranking (through games March 7)'], top_25_teams['FG%'])[0]*100
print(percent_correlation_FG_percent)
if (abs(percent_correlation_FG_percent) < 80):
    print("The correlation is not very strong")
else:
    print("The correlation is very strong")
```

# 3 Point Correlation

```python
top_25_teams["3sPerGame"] = top_25_teams["3P"]/top_25_teams["G"]
top_25_teams
```

```python
percent_correlation_3Pointers = stats.pearsonr(top_25_teams['Associated Press Ranking (through games March 7)'], top_25_teams['3sPerGame'])[0]*100
print(percent_correlation_3Pointers)
if (abs(percent_correlation_3Pointers) < 80):
    print("The correlation is not very strong")
else:
    print("The correlation is very strong")

print()
    
percent_correlation_3PointPercent = stats.pearsonr(top_25_teams['Associated Press Ranking (through games March 7)'], top_25_teams['3P%'])[0]*100
print(percent_correlation_3PointPercent)
if (abs(percent_correlation_3PointPercent) < 80):
    print("The correlation is not very strong")
else:
    print("The correlation is very strong")
    
```
